ans 2:

Generate your own annotation file and class names file.

One row for one image;

Row format: image_file_path box1 box2 ... boxN; (一張圖檔中，有哪幾個box。)

Box format: x_min,y_min,x_max,y_max,class_id (no space). (上述的每個box的詳細資訊，中間沒有空格。box與box之間需要空格。詳細請參考以下範例)

For VOC dataset, try python voc_annotation.py
